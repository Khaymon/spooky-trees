## Imports

```python
import pandas as pd
import sys

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

sys.path.append("/home/khaymonenko/trees-and-ensembles")

from trees_and_ensembles.models.spooky_tree import SpookyTree
```

## Data

```python
DATA_PATH = "/home/khaymonenko/trees-and-ensembles/data/competition/train.csv"
TEST_SIZE = 0.2

data = pd.read_csv(DATA_PATH)

X = data.drop(["class"], axis=1)
y = data["class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE)
```

```python
tree = SpookyTree(max_depth=8)
tree.fit(X_train, y_train)
```

```python
predictions = tree.predict(X_test)
accuracy_score(y_test, predictions)
```
