```python
import pandas as pd
```

```python
X = pd.read_csv("/home/khaymonenko/trees-and-ensembles/data/competition/train.csv", index_col="Id")
X
```

```python
cat_columns = ["does-bruise-or-bleed", "habitat", "season"]
mappings = {}

for feature in cat_columns:
    mappings[feature] = {value: idx for idx, value in enumerate(X[feature].unique())}
    X[feature] = X[feature].map(mappings[feature])

X.to_numpy()
```
